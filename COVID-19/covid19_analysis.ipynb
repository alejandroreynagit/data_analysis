### 4 - Compute mean, median and standard deviation for 3 of the features of the dataset. Use a for loop. What can you conclude with the information?
Don't forget the units of measurement.

```python
# Define list with features
features = ['Confirmed', 'Recovered', 'Deaths']

# Iterates over features and comnputes mean, median and standard deviation for each feature  
for i in features: 
    #covid19_ds[covid19_ds.columns[6]]
    print('Mean of', i , covid19_ds.loc[:,i].mean(), 'people')
    print('Median of', i , covid19_ds.loc[:,i].median(), 'people')
    print('Standard deviation of', i , covid19_ds.loc[:,i].std(), 'people\n')
```

### 6 - Calculate and visualise relations between variables, correlations. At least 3 plots, include interpretation.
All graphs should include title, axis names and units of measurement.

```python
covid19_ds2['Date'] = pd.to_datetime(covid19_ds2['Date'])
fig, ax = plt.subplots(figsize=(20,7))
plt.plot(covid19_ds2['Date'],covid19_ds2['Deaths'],'*r',markersize=9)
plt.plot(covid19_ds2['Date'],covid19_ds2['Confirmed'],'--y',linewidth=7)
plt.plot(covid19_ds2['Date'],covid19_ds2['Recovered'],'+g',markersize=10)
plt.title('Progression of Covid-19 cases worldwide',fontsize=20)
plt.xlabel('Date',fontsize=15)
plt.ylabel('Nr of people',fontsize=15)
plt.legend(fontsize=15)

# format xaxis with 4 month intervals
ax.get_xaxis().set_major_locator(mdates.DayLocator(interval=5))
ax.get_xaxis().set_major_formatter(mdates.DateFormatter("%d %b %Y"))
plt.setp(ax.get_xticklabels(), rotation=30, ha="right")
plt.grid()
plt.show()
```

* This is an accumulative sum trendline for all the confirmed cases, deaths and recoveries. We can see the progression of the impact regarding the Covid-19. The is an important number of confirmed cases that after receiving attention, showed a recovery. The number of deaths is increasing.

```python
fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(17,5))
covid19_ds.plot.scatter('Confirmed','Recovered', title = 'Recovered people out of confirmed cases', ax = axes[0], c='g',grid=True)
covid19_ds.plot.scatter('Confirmed','Deaths', title = 'Deaths out of confirmed cases', ax = axes[1],c='r', grid=True)
covid19_ds.plot.scatter('Recovered','Deaths', title = 'Recovered people vs. deaths', ax = axes[2],c='y',grid=True)
```

- A large number of confirmed cases have recovered after some time. The number of deaths keep increasing as the number of confirmed cases increase worldwide. It seems the world is taking this "seriously".

```python
# generate histograms
covid19_ds.hist(edgecolor='black', linewidth=1.9,figsize=(16,10));
```

- Histograms show the distribution of reference where the Covid-19 has a presence. The rest of the features are scalar values.

```python
covid19_ds.groupby('Country/Region')['Confirmed'].max().sort_values()[-20:].plot(kind='barh', figsize=(15,4),fontsize=10)
plt.xlabel('Nr of people')
plt.grid()
```

- From the distribution plot we can see clearly that USA (300k+) contributes with the highest number of confirmed cases of Covid-19, by far. Other countries such as Spain, Italy have similar numbers, followed by Germany and France. China appears to have recovered after being in the first place for weeks.
